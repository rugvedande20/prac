```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
for column1 in ['survived', 'sex', 'age', 'embarked', 'class', 'alive', 'alone']:
    for column2 in ['survived', 'sex', 'age', 'embarked', 'class', 'alive', 'alone']:
        if column1 == column2:
            break
        else:
            sns.countplot(data=df, x=column1, hue=column2)
            plt.title(f"{column1} V/S {column2}")
            plt.xlabel(f"{column1}")
            plt.ylabel(f"{column2}")
            plt.show()
```

```python
sns.histplot(df['fare'], bins=20, kde=True, edgecolor='black', color='purple')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```
