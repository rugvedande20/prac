```python
data = {
    'age' : ['20-25', '20-25','35-40','45-50', '45-50'],
    'income' : [40000, 50000, 90000, 120000, 190000]
}
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
summary_stats = df.groupby('age')['income'].describe()
summary_stats
```

```python
mean = df.groupby('age')['income'].mean()
print("Mean:\n",mean)
median = df.groupby('age')['income'].median()
print("\nMedian:\n",median)
std = df.groupby('age')['income'].std()
print("\nStandard Deviation:\n", std)
minn = df.groupby('age')['income'].min()
print("\nMin:\n", minn)
maxx = df.groupby('age')['income'].max()
print("\nMax:\n", maxx)
listt = df.groupby('age')['income'].apply(list)
print("\nList:\n", listt)
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv")
df.head()
```

```python
df.describe()
```

```python
for column in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    print(f"\n{column}:\n{df.groupby('species')[column].describe()}")
```

```python
for column in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    print(f"\nFOR {column}:")
    mean = df.groupby('species')[column].mean()
    print("Mean:\n",mean)
    median = df.groupby('species')[column].median()
    print("\nMedian:\n",median)
    std = df.groupby('species')[column].std()
    print("\nStandard Deviation:\n", std)
    minn = df.groupby('species')[column].min()
    print("\nMin:\n", minn)
    maxx = df.groupby('species')[column].max()
    print("\nMax:\n", maxx)
    listt = df.groupby('species')[column].apply(list)
    print("\nList:\n", listt)
    print()
```

```python
setosa = df[df['species']=='setosa']
virginica = df[df['species']=='virginica']
versicolor = df[df['species']=='versicolor']
```

```python
def display(group, species):
    print(f"\n FOR {species}:")
    numeric_group = group.select_dtypes(include='number')
    print(numeric_group.describe())

display(setosa, 'setosa')
display(virginica, 'virginica')
display(versicolor, 'versicolor')
```
