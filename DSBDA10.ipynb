 Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("iris")
data
```

# FEATURE

```python
print("\n 1.FEATURE AND THEIR TYPES:")
print(data.dtypes)
print("\n Summary:")
for column in data.columns:
    dtype=data[column].dtype
    if dtype=="object":
        print(f"{column}:Nominal")
    else:
        print(f"{column}:Numeric")
```

# HISTOGRAM

```python
print("\n 2. Histogram for feature")
numeric_feature=data.select_dtypes(include="number").columns
data[numeric_feature].hist(figsize=(10,6),color="lightblue",edgecolor="black")
plt.suptitle("Histogram for feature")
plt.tight_layout()
plt.show()
```

# BOXPLOT

```python
print("\n 3 Boxplot of feature")
plt.figure(figsize=(10,6))
for i, feature in enumerate(numeric_feature,1):
    plt.subplot(2,2,i)
    sns.boxplot(y=data[feature],color="lightblue")
    plt.title(f"Boxplot of {feature}")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data["sepal_length"])
```

```python
sns.boxplot(data=data,x="species",y="petal_length",palette="pastel")
```

# OUTLIERS

```python
print("\n 4 Outliers")
for feature in numeric_feature:
    Q1=data[feature].quantile(0.25)
    Q3=data[feature].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5 * IQR
    upper=Q3+1.5 * IQR
    outlier=data[(data[feature]<lower)|(data[feature]>upper)]
    print(f"{feature}:{len(outlier)} dected")
    
    
    
    
```

```python
sns.boxplot(data=data ,y="sepal_width",hue="species",palette="pastel")
```
